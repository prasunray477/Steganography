```python
import numpy as np
```

```python
d={}
c={}
for i in range(256):
  d[chr(i)]=i
  c[i]=chr(i)

text="Hello"
key="123"
```

```python
x=np.random.randint(0,255,(10,10,3),dtype=np.uint8)
x
```

```python
x_enc=x.copy()
n,m,z=0,0,0
l=len(text)
kl=0
for i in range(l):
  char_val=d[text[i]]^d[key[kl]]
  for bit_pos in range(8):
    bit=(char_val>>(7-bit_pos))&1
    org_val=x_enc[n,m,z]
    x_enc[n,m,z]=(org_val & 254)|bit
    z=(z+1)%3
    if z==0:
      m+=1
      if m==x_enc.shape[1]:
        m=0
        n=n+1
  kl=(kl+1)%len(key)
```

```python
kl,n,m,z=0,0,0,0
decrypt=""
for i in range(l):
  val=0
  for bit_pos in range(8):
    bit=x_enc[n,m,z]&1
    val=(val<<1)|bit
    print(f"Reading Bit{bit} from ({n},{m},{z})")
    z=(z+1)%3
    if z==0:
      m+=1
      if m==x_enc.shape[1]:
        m=0
        n=n+1
  decrypt+=c[val^d[key[kl]]]
  kl=(kl+1)%len(key)
print("Decrypt text",decrypt)
```
